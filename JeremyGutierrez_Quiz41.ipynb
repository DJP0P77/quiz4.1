### Please clean up the book title and save the results to a new dataframe table

```python
def clean_title(title):
    
    if title == 'nan':
        return 'NaN'
    
    if title[0] == '[':
        title = title[1: title.find(']')]
        
    if 'by' in title:
        title = title[:title.find('by')]
    elif 'By' in title:
        title = title[:title.find('By')]
        
    if '[' in title:
        title = title[:title.find('[')]

    return title

#Get the clean title
df['Title'] = df['Title'].apply(clean_title)
df.head()
```

### Example for .join()
#### DataFrame.join() is a convenient method for combining the columns of two potentially differently-indexed DataFrames into a single result DataFrame. 

```python
import pandas as pd

# Sample data: part 1
data1 = {
    'Country': ['USA', 'China', 'Russia'],
    'Gold': [39, 38, 20],
    'Silver': [41, 32, 28]
}
olympics_part1_df = pd.DataFrame(data1)
olympics_part1_df.set_index('Country', inplace=True)  # join works best on index

# Sample data: part 2
data2 = {
    'Country': ['USA', 'China', 'Russia'],
    'Bronze': [33, 18, 23],
    'Total': [113, 88, 71]
}
olympics_part2_df = pd.DataFrame(data2)
olympics_part2_df.set_index('Country', inplace=True)

# Join the two DataFrames
merged_df = olympics_part1_df.join(olympics_part2_df)

# Display the result
print(merged_df)
```

### In class Practice: 
1. Please use .merge or .join combine olympics data part1, part2, part3
2. Calculate the total number of accumulated medals of Summer and Winter Games

```python
import pandas as pd

# Read CSVs
columns = ['Country', 'Season', 'Gold', 'Silver', 'Bronze', 'Total']
part1 = pd.read_csv('olympics_part1.csv', header=None, names=columns)
part2 = pd.read_csv('olympics_part2.csv', header=None, names=columns)
part3 = pd.read_csv('olympics_part3.csv', header=None, names=columns)

# Convert 'Total' columns to numeric
part1['Total'] = pd.to_numeric(part1['Total'], errors='coerce')
part2['Total'] = pd.to_numeric(part2['Total'], errors='coerce')
part3['Total'] = pd.to_numeric(part3['Total'], errors='coerce')

# Merge Summer parts
summer = part1.merge(part2, on='Country', suffixes=('_part1', '_part2'))

# Merge with Winter part
full = summer.merge(part3, on='Country', suffixes=('', '_winter'))

# Sum totals
total_summer = full['Total_part1'].sum() + full['Total_part2'].sum()
total_winter = full['Total'].sum()
overall_total = total_summer + total_winter

# Display
totals_df = pd.DataFrame({
    'Summer_Total_Medals': [total_summer],
    'Winter_Total_Medals': [total_winter],
    'Overall_Total_Medals': [overall_total]
})

totals_df
```
